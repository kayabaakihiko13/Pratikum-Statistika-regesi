<div>
    <h1 align="center">Bab 2 Metode Korelasi </h1>
    <p align="justify">Analisis Korelasi Merupakan salah satu metode statiska yang mengelidiki hubungan dua variabel.Metode ini digunakan untuk mengetahui seberapa kuat atau lemahnya hubungan antar dua variabel.<br>contoh pengaplikasinya metode korelasi misalnya adalah kita mengetahui apakah varaibel ini memiliki hubungan kuat apa lemah (besti/tidak)</p>
</div>

```python
import pandas as pd
score_course=[90,80,85,65,70,65,95,82,83,75]
ipk=[3.85,3.72,3.82,3.2,3.4,3.1,3.9,3.75,3.76,3.49]
data=pd.DataFrame({
    "score":score_course,
    "ipk":ipk,
})
data.head()
```

<div>
    <h2 align="center">Pearsonr Method</h2>
</div>

```python
from scipy.stats import pearsonr
corr,p_value=pearsonr(data.score,data.ipk)
print('nilai coorelation is:{:.3f}'.format(corr))
print('nilai p-value:{}'.format(p_value))
```

<h3 align="center">Spearmanr Method</h3>

```python
from scipy.stats import spearmanr
corr,p_value=spearmanr(data.score,data.ipk)
print('nilai coorelation is:{:.3f}'.format(corr))
print('nilai p-value:{}'.format(p_value))
```

<h3 align="center">korelasi Tau Kendali</h3>

```python
from scipy.stats import kendalltau
corr,_=kendalltau(data.score,data.ipk)
print('nilai coorelation is:{:.3f}'.format(corr))
```

```python
# studi kasus
df=pd.read_csv('Data_Pendapatan.csv')
df=df.drop(columns=['Data ke-'],axis=1)
df.columns=df.columns.str.lower()
df.head()
```

```python
# bagaimana hunbungan dengan variabel bekerja_tahun dan pendapatan_juta
from statsmodels.api import qqplot
feature=df.bekerja_tahun
target=df.pendapatan_juta
fig,ax=plt.subplots(1,2,figsize=(20,10))
sns.scatterplot(x=feature,y=target,ax=ax[0])
ax[0].set_title('Coorlelation of bekerja_tahun and pendapatan_juta')
qqplot(target,line='s',ax=ax[1])
ax[1].set_title('Q-Q plot for pendapatan_juta')
plt.show()
```

```python
# pensebarannya sebelum modeling
sns.distplot(df.pendapatan_juta)
```

```python
# hubungan regesi sebelum di modeling
sns.regplot(x='bekerja_tahun',y='pendapatan_juta',data=df)
# ini masih berjauhan nah bagaimana untuk mengubah nya
```

Menbangun regesi


```python
# kita pakai pakage statsmodels
import statsmodels.api as sm
target_penjualan=df.pendapatan_juta
feature=df.bekerja_tahun
feature=sm.add_constant(feature)
modeling=sm.OLS(target_penjualan,feature).fit()# kita masukan (dependen,independen)
print(modeling.summary())
```

jadi B0 bernilai 2.579.000 dan untuk tahun b1=945.000
kesimpulannya bahwa setiap penganlaman kenaikan satu tahun akan bertambah pendapatan senilai 945.000

```python
# nah kita uji t_value dan p_value
plt.title('Pesebaran data setelah OLS')
sns.distplot(modeling.resid)
```

<div>
    <h4 align="center">cara melihat hasil prediksi dan residual</h4>
</div>

```python
nilai_prediksi=modeling.predict(feature)
print(nilai_prediksi.head())
```

```python
nilai_resedu=modeling.resid
print(nilai_resedu.head())
```

```python
from scipy.stats import shapiro
_,p_value=shapiro(modeling_iklan.resid)
if p_value>.05:
    print('Nilai H0 Tolak')
    print('Model Bagus')
else:
    print('Nilai Gagal H0 Tolak')
    print('Model Kurang Bagus')
print(p_value)
```

```python
# AutoKorelasi
print('nilai korelasi:',sm.stats.stattools.durbin_watson(model_data.resid))
```
